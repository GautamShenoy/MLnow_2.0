# Problem Statement 

Train your model for the equation : `y = 10 * x`

This is simple 10's table for which we are using machine learning (Missaplication)


#Step 1 

Data Creation : Creating your input (x = data, y = labels)
```
x = [0, 1, 2, 3, 4, ....]
y = [0, 10, 20, 30, 40, ....]
``` 

````python
x = [i for i in range(21)]
y = [i * 10 for i in range(21)]

````

#Step 2
Now we need to divide our data into training data and test data 

```
xTrain = [0, 1, 2, ..., 15]
yTrain = [0, 10, 20, ... , 50]

and 

xTest = [16, 17, .. , 20]
yTest = [160, ... , 200]
```



````python
xTrain = x[ : -5]
yTrain = x[ : -5]

xTest = x[-5 : ]
yTest = x[-5 : ]
````

````python
import tensorflow as tf
from tensorflow import keras
````

````python
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=1, input_shape = [1])
])
````

````python
model.compile(optimizer='adam', loss = 'mae')
````

````python
model.fit(x = xTrain, y = yTrain, validation_data = (xTest, yTest), epochs = 5000)
````
